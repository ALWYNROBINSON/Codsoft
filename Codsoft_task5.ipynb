```python
legit=d[d.Class==0]
fraud=d[d.Class==1]
```

```python
ls=legit.sample(n=492)
new_d=pd.concat([ls,fraud],axis=0)
new_d.head()
```

```python
new_d['Class'].value_counts()
```

```python
new_d.groupby('Class').mean()
```

```python
x=new_d.drop(columns='Class',axis=1)
y=new_d['Class']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_t,x_tst,y_t,y_tst=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
x.shape,x_t.shape,x_tst.shape
```

```python
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression()
mod.fit(x_t,y_t)
```

```python
ypl=mod.predict(x_tst)
LABELS=['Legit','Fraud']
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_tst,ypl)
plt.figure(figsize=(12,12))
sns.heatmap(cm,xticklabels=LABELS,yticklabels=LABELS,annot=True,fmt="d")
plt.title("Confusion Matrix")
plt.xlabel('predicted Class')
plt.ylabel('True Class')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
xtp=mod.predict(x_t)
tda=accuracy_score(xtp,y_t)
tda
```

```python
xtstp=mod.predict(x_tst)
tstda=accuracy_score(xtstp,y_tst)
tstda
```
